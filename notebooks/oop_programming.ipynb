```python
class duck:
    def walk(self):
        print("It walks like duck : ")
class horse:
    def walk(self):
        print("It walks like horse : ")
class cat:
    def talk(self):
        print("It talks like cat : ")

def myfunction(obj):
    if(hasattr(obj,'walk')):
        obj.walk()
    if(hasattr(obj,'talk')):
        obj.talk()
    
d=duck()
myfunction(d)
h=horse()
myfunction(h)
c=cat()
myfunction(c)
```

# method overriding 


```python
class add:
    def addition(self,a,b):
        print("addition :" , a+b)
        
class multi(add):
    
    def addition(self,a,b):
        super().addition(a,b)
        print("Multiplication :" , a*b)

m1=multi()
m1.addition(10,20)
print()
m2=multi()
m2.addition(100,200)
```

# abstract class

```python
from abc import ABC , abstractmethod

class defence_force(ABC):
    def gun (self):
        print("All the defence forces has AK47")
    @abstractmethod
    def area(self):
        pass
class Army(defence_force):
    def area(self):
        print("area of interest")
        print("Indian Land")
        
class IAF(defence_force):
    def area(self):
        print("area of interest")
        print("Indian Air Sapce")
        
class Navy(defence_force):
    def area(self):
        print("area of interest")
        print("Indian Ocean")
        
a1=Army()
a1.gun()
a1.area()
print()

a1=IAF()
a1.gun()
a1.area()
print()

a1=Navy()
a1.gun()
a1.area()
print()
```
